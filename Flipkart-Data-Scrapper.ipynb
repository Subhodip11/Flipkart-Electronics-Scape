```python
import requests
from bs4 import BeautifulSoup
from collections import defaultdict
import pandas as pd
```

```python
names=[]
prices=[]
details=[]
Data=[]
domain='https://www.flipkart.com'
for i in range(1,10):
    #print('Phone Details ',i)
    pageNo=str(i)
    URL=domain+'/search?q=smart+watches&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&page='+pageNo
    r = requests.get(URL)
    soup = BeautifulSoup(r.text,'html5lib')
    for rawdata in  soup.findAll(class_ = '_2kHMtA'):
        aTag = rawdata.find('a')
        Image = aTag.find('div',{'class' : 'CXW8mj'}).find('img')['src']
        #print(images)
        Info = rawdata.find('div',{'class':'_3pLy-c row'})
        Stats = Info.findAll('div')[0].findAll('div')
        #Phone Name
        Name = Stats[0].text
        #print(phoneName)
        #Phone price
        Price = Info.find('div',{'class':'_3tbKJL'}).find('div').find('div').text
        #Phone Details
        Details =Info.find('div',{'class':'fMghEO'}).text
        names.append(Name)
        prices.append(Price)
        details.append(Details)
        Data.append([Image,Name,Price,Details])
```
